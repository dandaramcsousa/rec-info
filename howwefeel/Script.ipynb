```python
import csv
import pandas as pd

import nltk as nl
from collections import defaultdict
```

```python
df = pd.read_csv("./howwefeel/db.csv", encoding='latin-1')
df.head()
```

```python
def term_freq(term, doc):
    return len(list(filter((lambda x: x == term), doc)))
```

```python
def gera_tokens(df):
    indexes = {}
    for index,row in df.iterrows():
        text_tokens = (word.lower() for word in (nl.word_tokenize(row['text'])))

        
        tokens = list(text_tokens)
        
        for token in tokens:
            if (token in indexes):
                indexes[token][token][row['id']] = term_freq(token,tokens)
            else:
                indexes[token] = {token: {row['id'] : term_freq(token,tokens)}}
            
    return indexes
```

```python
i_indexes = gera_tokens(df)
```

```python
def term_freq(query):
    termf_dict = {}
    for term in query:
        for doc_id in i_indexes[term][term].keys():
            if (doc_id in termf_dict):
                termf_dict[doc_id] += i_indexes[term][term][doc_id]
            else:
                termf_dict[doc_id] = i_indexes[term][term][doc_id]
    return termf_dict
```

```python
def binary_rep(query):
    bin_dict = {}
    for term in query:
        for doc_id in i_indexes[term][term].keys():
            if doc_id in bin_dict:
                bin_dict[doc_id] += 1
            else:
                bin_dict[doc_id] = 1
    return bin_dict
```
